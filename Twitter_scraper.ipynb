```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper = Nitter()
```

```python
def get_tweets(name, modes, no):
  tweets = scraper.get_tweets(name, mode = modes, number=no)
  final_tweets = []
  for tweet in tweets['tweets']:
    data = [tweet['link'],tweet['text'], tweet['date'], tweet['stats']['likes'], tweet['stats']['comments']]
    final_tweets.append(data)
  df = pd.DataFrame(final_tweets, columns=['link','text','date','likes','comments'])
  return df
```

```python
topic = input("Enter a topic you want tweets on : ")
data = get_tweets(topic, 'hashtag', 100)
```
